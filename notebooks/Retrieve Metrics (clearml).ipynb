# SBRI (_Selection Bias Robustness Index_)

```python
from sklearn.metrics import cohen_kappa_score
```

```python
# Baseline
pred_info_vit_baseline = pd.read_csv(
    os.path.join(BASE_DIR, 'baseline', 'evaluation', 'test_pred_info_vit.csv'))
pred_info_vit_baseline = pred_info_vit_baseline.iloc[:, 1:]

pred_info_bit_baseline = pd.read_csv(
    os.path.join(BASE_DIR, 'baseline', 'evaluation', 'test_pred_info_bit.csv'))
pred_info_bit_baseline = pred_info_bit_baseline.iloc[:, 1:]
# Spurious
pred_info_vit_spurious = pd.read_csv(
    os.path.join(BASE_DIR, 'spurious', 'evaluation', 'test_pred_info_vit.csv'))
pred_info_vit_spurious = pred_info_vit_spurious.iloc[:, 1:]

pred_info_bit_spurious = pd.read_csv(
    os.path.join(BASE_DIR, 'spurious', 'evaluation', 'test_pred_info_bit.csv'))
pred_info_bit_spurious = pred_info_bit_spurious.iloc[:, 1:]
```

## ViT

```python
df_pred_info_vit = pd.merge(
    pred_info_vit_baseline, pred_info_vit_spurious, 
    on='image', how='inner', suffixes=('_base', '_spur'))
```

```python
sbri_vit = cohen_kappa_score(df_pred_info_vit['pred_base'], df_pred_info_vit['pred_spur'], weights='quadratic')
```

```python
sbri_vit
```

```python
df_pred_info_vit_filtered = df_pred_info_vit.query('pred_base == label_base')
```

```python
sbri_vit_filt = cohen_kappa_score(
    df_pred_info_vit_filtered['pred_base'], df_pred_info_vit_filtered['pred_spur'], weights='quadratic')
```

```python
sbri_vit_filt
```

## BiT

```python
df_pred_info_bit = pd.merge(
    pred_info_bit_baseline, pred_info_bit_spurious, 
    on='image', how='inner', suffixes=('_base', '_spur'))
```

```python
sbri_bit = cohen_kappa_score(df_pred_info_bit['pred_base'], df_pred_info_bit['pred_spur'], weights='quadratic')
```

```python
sbri_bit
```

```python
df_pred_info_bit_filtered = df_pred_info_bit.query('pred_base == label_base')
```

```python
sbri_bit_filt = cohen_kappa_score(
    df_pred_info_bit_filtered['pred_base'], df_pred_info_bit_filtered['pred_spur'], weights='quadratic')
```
